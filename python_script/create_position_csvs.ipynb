```python
import mne
import csv
```

```python
mne.channels.montage.get_builtin_montages()
```

```python
for montage_name in mne.channels.montage.get_builtin_montages():
    print(montage_name)
    layout = mne.channels.make_standard_montage(montage_name)
    layout.plot()
```

```python
for montage_name in mne.channels.montage.get_builtin_montages():
    montage = mne.channels.make_standard_montage(montage_name)
    positions_dict = montage.get_positions()['ch_pos']
    positions = list(positions_dict.values())
    labels = list(positions_dict.keys())
    with open('../data/sensor_montages/'+montage_name+'_coordinates.csv', 'w+') as f:
        writer = csv.writer(f, delimiter=',')
        for row in positions[:-1]:
            writer.writerow(row)
    with open('../data/sensor_montages/'+montage_name+'_labels.csv', 'w+') as f:
        writer = csv.writer(f, delimiter=',')
        for row in labels[:-1]:
            writer.writerow(row)
```
