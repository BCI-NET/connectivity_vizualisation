```python
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
n_points = 50

x = []
y = []
z = []

for i in range(n_points):
    theta_i = 2*math.pi * np.random.rand()

    z_ = 2 * np.random.rand() - 1
    y_ = math.sqrt(1 - z_**2) * math.cos(theta_i)
    x_ = math.sqrt(1 - z_**2) * math.sin(theta_i)

    x.append(x_)
    y.append(y_)
    z.append(z_)

u = np.linspace(0, 2* np.pi, 100)
v = np.linspace(0, np.pi, 100)
x_sphere = np.outer(np.cos(u), np.sin(v))
y_sphere = np.outer(np.sin(u), np.sin(v))
z_sphere = np.outer(np.ones(np.size(u)), np.cos(v))
```

```python
%matplotlib notebook
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x,y,z, color='r')
ax.plot_surface(x_sphere, y_sphere, z_sphere, rstride=8, cstride=8, alpha=0.3)
```

```python
coords = []
for i in range(len(x)):
    coords.append([x[i], y[i], z[i]])
```
